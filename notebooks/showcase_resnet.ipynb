# ResNet50 Architecture Showcase

This notebook demonstrates how to load the ResNet50 architecture using TensorFlow/Keras and explore its layer structure.

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50

# Load the ResNet50 model
# include_top=True includes the final Dense layers for ImageNet classification
model = ResNet50(weights='imagenet', include_top=True)

# Display the summary of the model
model.summary()
```

```python
# Detailed view of the first 20 layers
print(f"{'Layer Name':<30} | {'Layer Type':<20}")
print("-" * 55)
for layer in model.layers[:50]:
    print(f"{layer.name:<30} | {type(layer).__name__:<20}")
```
